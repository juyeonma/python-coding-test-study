## 풀이방법
- 순서대로 조건문 쓰면 끄읕
- 사방의 테두리에는 벽이 있고, 벽에는 청소기가 갈 수 없으므로 index 조건 체크 안해도 된다.
- 반시계 방향으로 회전할 때, (d-1) % 4 = (d+3) % 4. 그렇지만 후자가 더 빠르다.
  
1. 재귀함수 이용
- `sys.exit(0)`을 사용하면 프로그램을 바로 종료하므로, 반드시 먼저 `print(answer)`을 해줘야한다.

2. while 이용
- 이게 쪼끔 더 빠르다!

## 코드
### 1. 재귀함수 이용

```python
import sys
input = sys.stdin.readline

# n * m 크기의 방
n, m = map(int, input().split())
# 처음 r, c 좌표, 방향 d
r, c, d = map(int, input().split())

# 0: 빈칸, 1: 벽
room = [list(map(int, input().split())) for _ in range(n)]
    
# 방향: 북 동 남 서 -> 반시계 방향 회전: (index + 3) % 4
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

def cleaning(x, y, d):
    global answer
    
    # 1. 
    if room[x][y] == 0:
        room[x][y] = 2
        answer += 1
        
    # 3.
    for _ in range(4):
        d = (d + 3) % 4
        nx = x + dx[d]
        ny = y + dy[d]
        if room[nx][ny] == 0:
            cleaning(nx, ny, d)
            return
    # 2.
    # 방향: 북 -> 뒤로 한칸. 즉 남으로 1칸 + 함.
    nx = x - dx[d]
    ny = y - dy[d]
    if room[nx][ny] == 1:
        print(answer)
        sys.exit(0)
        
    # 벽이 아니라면, 후진    
    cleaning(nx, ny, d)

answer = 0
cleaning(r, c, d)
```

### 2. while 이용: 이게 더 빠르다!

```python
import sys
input = sys.stdin.readline

# n * m 크기의 방
n, m = map(int, input().split())
# 처음 r, c 좌표, 방향 d
r, c, d = map(int, input().split())

# 0: 빈칸, 1: 벽
room = [list(map(int, input().split())) for _ in range(n)]
    
# 방향: 북 동 남 서 -> 반시계 방향 회전: (index + 3) % 4
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

def cleaning(x, y, d):
    # 1.
    room[x][y] = 2
    answer = 1

    while True:
        # 3.
        for _ in range(4):
            d = (d + 3) % 4
            nx = x + dx[d]
            ny = y + dy[d]
            if room[nx][ny] == 0:
                room[nx][ny] = 2
                answer += 1
                x, y = nx, ny
                break
        # 2.
        # 방향: 북 -> 뒤로 한칸. 즉 남으로 1칸 + 함.
        else:
            x -= dx[d]
            y -= dy[d]
            if room[x][y] == 1:
                return answer

print(cleaning(r, c, d))
```
