### 풀이 방법
**규칙을 찾지 못해서, 결국 답 검색.**
- 뒤에서부터, 연속된 두 숫자 중 앞에 것이 더 큰 것을 a로 한다.
- 다시 뒤에서부터, a보다 큰 숫자를 찾아 b로 한다.
- a와 b의 값을 서로 바꾼다.
- a 이후의 숫자들의 순서를 정렬한다(즉, 뒤집는다).
```
SHUTTLE: 4265531

2 < 6? YES
2 < 3? YES
4365521 -> 4312556: SLEHTTU
```

### 코드
- for-else 문을 활용하자.

````python
import sys
input = sys.stdin.readline

def word(s):
    len_s = len(s)
    for a in range(len_s-2, -1, -1):
        if s[a] < s[a+1]:
            break
    else: # s[a] >= s[a+1]
        return ''.join(s)

    for b in range(len_s-1, -1, -1):
        if s[a] < s[b]:
            break

    s[a], s[b] = s[b], s[a]
    s[a+1:] = sorted(s[a+1:])

    return ''.join(s)
    
for _ in range(int(input())):
    s = list(input().rstrip())
    print(word(s))
````
