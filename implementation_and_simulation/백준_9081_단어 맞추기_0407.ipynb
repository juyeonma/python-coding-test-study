- for-else  .

````python
import sys
input = sys.stdin.readline

def word(s):
    len_s = len(s)
    for a in range(len_s-2, -1, -1):
        if s[a] < s[a+1]:
            break
    else: # s[a] >= s[a+1]
        return ''.join(s)

    for b in range(len_s-1, -1, -1):
        if s[a] < s[b]:
            break

    s[a], s[b] = s[b], s[a]
    s[a+1:] = sorted(s[a+1:])

    return ''.join(s)
    
for _ in range(int(input())):
    s = list(input().rstrip())
    print(word(s))
````
